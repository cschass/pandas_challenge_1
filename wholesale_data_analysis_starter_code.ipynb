## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
filtered_df = client_df[client_df['client_id'].isin([66037, 33615]).copy()]
filtered_df.loc[:, 'subtotal'] = filtered_df['unit_price'] * filtered_df['qty']
result_df = filtered_df[['unit_price', 'qty', 'subtotal']]
result_df = filtered_df[["unit_price", "qty", "subtotal"]]
result_df.reset_index(drop=True, inplace=True)

print(result_df)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def calculate_shipping_price(row):
        if row['total_weight'] > 50:
            return row['total_weight'] * 7
        else:
            return row['total_weight'] * 10
client_df['total_weight'] = client_df['unit_weight'] * client_df['qty']
client_df['shipping_price'] = client_df.apply(calculate_shipping_price, axis=1)

print(client_df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head())
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
client_df['line_subtotal'] = client_df['unit_price'] * client_df['qty']
client_df['line_price'] = (client_df['line_subtotal'] + client_df['shipping_price']) * 1.0925

print(client_df[['line_subtotal', 'shipping_price', 'line_price']].head())
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
client_df['line_cost'] = client_df['unit_cost'] * client_df['qty'] + client_df['shipping_price']

client_df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
client_df['line_cost'] = client_df['unit_cost'] * client_df['qty'] + client_df['shipping_price']
client_df['line_price'] = client_df['unit_price'] * client_df['qty']
client_df['profit'] = client_df['line_price'] - client_df['line_cost']
client_df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

top_5_clients = [33615, 66037, 46820, 38378, 24741]
total_spending = {}
for client_id in top_5_clients:
    total_spending[client_id] = client_df[client_df['client_id'] ==client_id]['total_price'].sum()
    print(f"Total spending for client {client_id}: ${total_spending[client_id]:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

summary_data = {
    'client_id':[],
    'total_units_purchased': [],
    'total_shipping_price': [],
    'total_revenue': [],
    'total_profit': []
}
for client_id in top_5_clients:
    client_data = client_df[client_df['client_id'] == client_id]
    total_units = client_data['qty'].sum()
    total_shipping = client_data['shipping_price'].sum()
    total_revenue = client_data['total_price'].sum()
    total_profit = client_data['profit'].sum()
    summary_data['client_id'].append(client_id)
    summary_data['total_units_purchased'].append(total_units)
    summary_data['total_shipping_price'].append(total_shipping)
    summary_data['total_revenue'].append(total_revenue)
    summary_data['total_profit'].append(total_profit)
    summary_df = pd.DataFrame(summary_data)
    print(summary_data)
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_columns =['total_shipping_price', 'total_revenue', 'total_profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(amount):
    return f"$(amount/1,000,000:.2f)M"

# Apply the currency_format_millions function to only the money columns. 
for column in money_columns:
    summary_df[column] = summary_df[column].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
summary_df.rename(columns={
    'total_shipping_price': 'Total Shipping Price (Millions)',
    'total_revenue': "Total Revenue (Millions)",
    'total_profit': "Total Profit (Millions)"
})
print(summary_df)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_summary_df = summary_df.sort_values(by='Total Profit (Millions)',ascending=False)
print(sorted_summary_df)
```
